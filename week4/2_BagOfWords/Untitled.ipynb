```python
import nltk 
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
import re

paragraph = """
The news mentioned here is fake. Audience do not encourage fake news. Fake news is false or misleading.
People should be careful when reading news online. Not all information on the internet is true.
Some websites spread fake news to gain attention. Readers must verify sources before believing anything.
Fake news can create confusion in society. It can also cause unnecessary panic.
Many people share news without checking facts. This behavior helps fake news spread faster.
Social media platforms are full of mixed information. Some are true, some are completely false.
It is important to think before sharing any news. Critical thinking helps avoid misinformation.
Fake news often uses attractive headlines. These headlines trick readers into clicking.
Once people believe fake news, it becomes hard to correct. Misinformation spreads quickly.
Educated readers tend to question information. They look for trusted sources.
Fake news can harm individuals and communities. It can damage reputations.
Some fake news is created intentionally. Others are spread unknowingly.
Technology makes it easy to share information. But it also makes spreading fake news easier.
Users must be responsible online. They should not trust everything they see.
News should come from reliable sources. Trusted journalism is important.
Fake news can influence public opinion. It can even affect decisions.
People should cross-check information. One source is not enough.
Images and videos can also be misleading. They can be edited easily.
Fake news often plays with emotions. It makes people react quickly.
Staying calm helps in judging information. Emotional reactions can mislead.
Students should learn media literacy. It helps them identify fake content.
Fake news is a growing problem worldwide. Many countries are affected.
Governments try to control misinformation. But it is not always easy.
Everyone has a role in stopping fake news. Awareness is the first step.
Reading carefully can prevent mistakes. Skimming leads to misunderstanding.
Fake news sometimes looks very real. It can be difficult to detect.
Checking the author is important. Unknown sources are risky.
Dates also matter in news. Old news can be misleading if reused.
Fake news spreads faster than real news. This is a serious issue.
Algorithms sometimes promote trending content. Even if it is false.
Users should report fake news. Platforms often provide this feature.
Discussion with others can help verify news. Different views are useful.
Fake news can target specific groups. It can create division.
Awareness campaigns help educate people. They reduce misinformation.
Schools should teach critical thinking. It is a necessary skill.
Fake news affects trust in media. People become more skeptical.
Reliable journalism builds credibility. It takes time and effort.
Readers must take responsibility. Passive reading is dangerous.
Fake news can impact mental health. It creates fear and anxiety.
Information overload makes it harder to judge truth. Too much data confuses people.
Simple verification steps can help. Checking multiple sources is effective.
Fake news is not always obvious. It can hide in plain sight.
People should slow down when reading. Rushing leads to errors.
Understanding context is important. Misinterpretation causes problems.
Fake news thrives on ignorance. Knowledge is the best defense.
Being informed is powerful. It protects against misinformation.
Truth should always be valued. It builds a better society.
"""
sentences = nltk.sent_tokenize(paragraph)
lemmatizer = WordNetLemmatizer()
corpus = [] 
for i in range(len(sentences)):
    sent = re.sub('[^a-zA-Z]', ' ', sentences[i])
    sent = sent.lower()
    sent= sent.split()
    sent = [lemmatizer.lemmatize(word) for word in sent if not word in set(stopwords.words('english'))]
    sent = ' '.join(sent)
    corpus.append(sent)
print(corpus)
```

```python
#List all unique words
print("corpus:", corpus)
```

```python
#Finding unique words 
all_words = [] 
for sentence in corpus : 
    for word in sentence.split():
        all_words.append(word)
unique_words = list(set(all_words))
print("Uniwue words:", unique_words)
```

```python
# Create a dictionary with mapping of words to a number. This should now be sorted on frequency of occurrence in descending order.
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

word_frequencies = Counter(all_words)

# Print the word counts
print("Word Frequencies:", word_frequencies)

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print("Word to Number Mapping (Sorted by Frequency):", word_to_number)
```
