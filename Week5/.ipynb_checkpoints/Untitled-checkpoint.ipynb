```python
#coisine similarities
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity 
from bs4 import BeautifulSoup
import requests
```

```python
data = "https://bnpb.go.id/berita/banjir-setinggi-1-5-meter-di-kota-ambon-kini-berangsur-surut"
```

```python
def scrap_data(url): 
    contents = []
    try: 
        response = requests.get(url)
        if response.status_code == 404: 
            contents.append({"message": "can't access"})
        else : 
            soup = BeautifulSoup(response.content, 'html.parser')
            title = soup.find(class_="title").text
            paragraph = soup.find_all("p")
            content = "\n".join([p.text for p in paragraph])
            contents.append({"title":title, "content" : content })
            
    except Exception as e: 
        contents.append(str(e))
    return contents
    
```

```python
result = scrap_data(data)
print(result)
```

```python
df = pd.DataFrame(result)
df.head()
```

```python
content = df["content"][0]
print(content)
```

```python
import re
def elimination(text):
     # Remove date/view count lines (start with digit + month + year)
    text = re.sub(r'^\s*\d{1,2}\s\w+\s\d{4}.*$', '', text, flags=re.MULTILINE)
    
    # Remove "Dilihat ... kali" lines
    text = re.sub(r'^\s*Dilihat.*$', '', text, flags=re.MULTILINE)
    
    # Remove photo captions (lines containing "Foto :")
    text = re.sub(r'^\s*Foto\s*:.*$', '', text, flags=re.MULTILINE)
    
    # Remove footer lines (Telp, Fax, Email, phone numbers, names)
    text = re.sub(r'^\s*(Telp.*|Fax.*|Email.*|\+?\d[\d\s]+)$', '', text, flags=re.MULTILINE)
    
    # Collapse multiple blank lines
    text = re.sub(r'\n+', '\n', text)
    return text.strip()
    
```

```python
df["newcontent"] = df["content"].apply(elimination)
```

```python
print(df["newcontent"][0])
```

```python
import re

paragraphs = df["newcontent"][0]

# Split into sentences
sentences = re.split(r'\.\s*', paragraphs)

# Clean up and add back the period
sentences = [s.strip() + '.' for s in sentences if s.strip()]

# Now sentences is already a list
print(sentences)
```

```python
vectorizer = CountVectorizer(vocabulary=['banjir', 'warga'])
x = vectorizer.fit_transform(sentences)
print(x.toarray())
```

```python
#Count the Cosine Similarity 
con_sim = cosine_similarity(x[0:1],x)

#showing the result
df = pd.DataFrame(x.toarray(), columns=['banjir','warga'])
df['Cosine Sim to Doc 0'] = con_sim.flatten()
print(df)
```

```python
from collections import Counter 
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in sentences],).fillna(0)
tf
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```

```python
full_cos_sim_matrix = cosine_similarity(x)

# Display the full matrix as a DataFrame for better readability
cos_sim_df = pd.DataFrame(full_cos_sim_matrix, index=[f'Doc {i}' for i in range(len(sentences))], columns=[f'Doc {i}' for i in range(len(sentences))])
print("Full Cosine Similarity Matrix:")
print(cos_sim_df)
```
