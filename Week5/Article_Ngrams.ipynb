```python
import requests
from bs4 import BeautifulSoup

def scrap_data(data):
    contents = []

    for i in data:
      try:
        response = requests.get(i)
        soup = BeautifulSoup(response.content, "html.parser")
        title = soup.find("title").text
        paragraphs = soup.find_all("p")
        content = "\n".join([p.text for p in paragraphs])
        contents.append({"title": title,"url": i ,"content": content})

      except Exception as e:
        contents.append(str(e))

    return contents



```

```python
result = scrap_data(data)
print(result)
```

```python
df = pd.DataFrame(result)#can used for list or dict
df.to_csv("data.csv", index=False)
```

```python
df.head()
```

```python
df_content1 = df["content"][0]
ngram_object = TextBlob(df_content1)
print(ngram_object)
```

```python
ngrams = ngram_object.ngrams(n=2)
print(ngrams)
```

```python
#Tri-Grams
ngram_object = TextBlob(df_content1)
ngrams3 = ngram_object.ngrams(n=3)
print(ngrams3)
```
