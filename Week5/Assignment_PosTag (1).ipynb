```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrap_text(data):
    results = []

    for i in data:
        try:
            response = requests.get(i)
            soup = BeautifulSoup(response.content, "html.parser")

            title_tag = soup.find('title')
            title = title_tag.text if title_tag else "No title"

            paragraphs = soup.find_all("p")
            content = "\n".join([para.text for para in paragraphs])

            results.append({
                "title": title,
                "url": i,
                "content": content
            })

        except Exception as e:
            results.append({
                "title": "error",
                "url": i,
                "content": str(e)
            })

    return results
```

```python
pos2 = pos_tag(token2)
print(pos2)
```

```python
print("{:<16}{}".format("Word","Pos tag"))
print("-"*30)

for word, tag in pos2:
  print("{:<16}{:>2}".format(word, tag))
```

```python
from collections import Counter
word_pos_list = [{"Word": word, "Pos Tag" : tag} for word, tag in pos2]

df_word_pos = pd.DataFrame(word_pos_list)

tag_counts = df_word_pos['Pos Tag'].value_counts()

unique_token_per_tag = df_word_pos.groupby('Pos Tag')['Word'].nunique()

summary_df = pd.DataFrame({
    'Tag' : tag_counts.index,
    'Count' : tag_counts.values,
    'Unique Tokens': unique_token_per_tag.values
}).reset_index(drop=True)

summary_df = summary_df.sort_values(by='Count', ascending=False)
print(summary_df)
summary_df.to_csv("pos_tags_summary.csv",index=False)
```
